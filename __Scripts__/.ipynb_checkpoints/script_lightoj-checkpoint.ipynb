```python
volume = 10
page = 1

driver.get(f"https://lightoj.com/problems/category/loj-vol-{volume}?page={page}")
time.sleep(10)
```

```python
problem_links = []
problem_id = []

table = driver.find_elements(By.CSS_SELECTOR, '.page-block')
for row in table:
    col = row.find_element(By.CSS_SELECTOR, '.is-pulled-left')
    if col.text=="Solve":
        continue
    link = row.find_element(By.TAG_NAME, 'a')
    problem_links.append(link.get_attribute("href"))
    
    _id = row.find_element(By.CSS_SELECTOR, '.loj-id')
    problem_id.append(_id.text[4:])
```

```python
from bs4 import BeautifulSoup
#soup = BeautifulSoup(driver.getPageSource())
soup = BeautifulSoup(driver.page_source, 'lxml')
print(soup.prettify())
```

```python
mp = {
    "\\u0021": "!",
    "\\u0022":  "\"",
    "\\u0023":  "#",
    "\\u0024":  "$",
    "\\u0025":  "%",
    "\\u0026":  "&",
    "\\u0027":  "'",
    "\\u0028":  "(",
    "\\u0029":  ")",
    "\\u002A":  "*",
    "\\u002B":  "+",
    "\\u002C":  ",",
    "\\u002D":  "-",
    "\\u002E":  ".",
    "\\u002F":  "/",
    "\\u003A":  ":",
    "\\u003B":  ";",
    "\\u003C":  "<",
    "\\u003D":  "=",
    "\\u003E":  ">",
    "\\u003F":  "?",
    "\\u0040":  "@",

}

```

```python
import unicodedata
def process(s):
    ret = ""
    
    ii = s.find("#include")
    if ii==-1:
        return ret
    
    jj = s.rfind("return 0")
    
    ret = s[ii: jj+14]
    ret = ret.replace("\\n", "\n")
    ret = ret.replace("\\r", "")
    ret = ret.replace("\\\"", "\"")
    
    for x, y in mp.items():
        ret = ret.replace(x, y)
    
    #ret = ret.replace("\\u003C", '<')
    #ret = ret.replace("\\u002F", '>')
    
    print(ret)


code = soup.find_all('script')
final = ""
for _ in code:
    s= str(_.text)
    process(s)
    
#with open("tmp.cpp", "w") as f:
#    f.write(final)
#    f.close()
```
